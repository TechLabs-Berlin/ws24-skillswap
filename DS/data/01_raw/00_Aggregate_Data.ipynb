```python
import pandas as pd
```

```python
Personal_Info = pd.read_csv('Personal_Info.csv')
Skill_Exchange = pd.read_csv('Skill_Exchange.csv')
Track_Record = pd.read_csv('Track_Record.csv')
Feedback = pd.read_csv('Feedback.csv')


# Concatenate DataFrames horizontally
combined_df = pd.concat([Personal_Info, Skill_Exchange, Track_Record, Feedback], axis=1)

combined_df.head(750) 
```

```python
combined_df.to_csv('Skillswap_Data_Raw.csv', index=False)
```
