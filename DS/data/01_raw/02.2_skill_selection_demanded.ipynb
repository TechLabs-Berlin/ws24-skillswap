```python
import random
import pandas as pd
import numpy as np
```

```python
# Define the skill domains
# Pick a random number between 1 and 5
# Select random skill domains
# Generate 750 random skill domain samples


skill_domains = [
    "Professional_Skills",
    "Creative_Skills",
    "Language_Skills",
    "Technical_Skills",
    "Social_Skills",
    "Practical_Skills",
    "Academic_Skills",
    "Life_Skills"
]

def select_random_skills():
    num_skills = random.randint(1, 5)

    selected_skills = random.sample(skill_domains, num_skills)
    
    return selected_skills

random_skill_samples = []

for _ in range(750):
    random_skills = select_random_skills()
    random_skill_samples.append(random_skills)
    
print(random_skill_samples)
```

```python
# Initialize a list to store the matched skills for each domain sample
# Iterate through each domain sample in random_skill_samples
# Initialize an empty list to store the matched skills for the current sample
# Iterate through each skill domain in the current sample
# Retrieve the matching skills from skills_dict and extend the matched_skills list
# Append the matched skills for the current sample to the matched_skills_samples list


matched_skills_samples = []

for sample in random_skill_samples:
    matched_skills = []
    for skill_domain in sample:
        matched_skills.extend(skills_dict.get(skill_domain, []))
    matched_skills_samples.append(matched_skills)
    
    
print(matched_skills_samples)
```

```python
concatenated_df = pd.concat([Skills_Demanded, Matched_skills_df], axis=1)
concatenated_df
```

```python
# Randomly select a value from possible_levels with a normal distribution to 
# fill the columns (where a person has a skill interest) with skill levels
# Define possible skill levels
# Assumption: Usually people tend to seek for people that are rather experienced to teach them skills
# Iterate over each row in the DataFrame
# Extract the list of matched skills from the current row
# Iterate over each column in the DataFrame
# Check if the column name matches any of the skills in matched_skills
# Define parameters for the normal distribution
# Generate random index from a normal distribution
# Ensure the index falls within the range of possible indices
# Select the skill level corresponding to the generated index
# Assign the randomly selected value to the corresponding cell in the matched headline

possible_levels = ["Novice", "Beginner", "Intermediate", "Advanced", "Expert"]

average_range = [possible_levels.index("Advanced"), possible_levels.index("Expert")]

for index, row in concatenated_df.iterrows():
    matched_skills = row['Matched_skills_samples']
    
    for column in concatenated_df.columns:
        if column in matched_skills:
            mean = np.mean(average_range)  
            std_dev = len(possible_levels) / 3  
            
            random_index = int(np.random.normal(mean, std_dev))
            random_index = max(0, min(random_index, len(possible_levels) - 1))
            random_value = possible_levels[random_index]

            concatenated_df.at[index, column] = random_value
```
