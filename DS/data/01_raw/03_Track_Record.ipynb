```python
import pandas as pd
import random
```

```python
# Create lists containing the number of skills offered and taken
Count_Skills_Off = []
Count_Skills_Rec = []
Rec_to_Off_Ratio = []
```

```python
# Ensure a good ratio between skills offered and received


# Calculate the maximum deviation allowed for a given ratio.
def calculate_max_deviation(count_received, max_deviation_percent):
    # Calculate the maximum allowable deviation based on received skills
    max_deviation = count_received * max_deviation_percent / 100.0
    return max_deviation

# Define parameters
# try to keep the ratio at 1:1 skill offered(received) with a maximum 
# allowable deviation percentage of 20% 
max_deviation_percent = 20
max_count = 300  
num_iterations = 750

# Generate random skills received and calculate skills offered within a deviation range
for _ in range(num_iterations):
    # Generate random count of skills received
    count_received = random.randint(1, max_count)
    
    # Calculate the maximum deviation allowed for skills offered based on received skills
    max_deviation = calculate_max_deviation(count_received, max_deviation_percent)
    
    # Generate count of skills offered within the deviation range
    count_offered = random.randint(max(1, count_received - round(max_deviation)), min(max_count, count_received + round(max_deviation)))
    
    # Append values to the lists
    Count_Skills_Rec.append(count_received)
    Count_Skills_Off.append(count_offered)
    
    # Calculate and append the ratio
    ratio = round(count_received / count_offered, 2)
    Rec_to_Off_Ratio.append(ratio)
```

```python
# Create a data frame
table = {
    'Count_Skills_Rec': Count_Skills_Rec,
    'Count_Skills_Off': Count_Skills_Off,
    'Rec_to_Off_Ratio': Rec_to_Off_Ratio
}

# Create a DataFrame
Track_Record = pd.DataFrame(table)
Track_Record.head(100)
```

```python
Track_Record.to_csv('Track_Record.csv', index=False)
```
