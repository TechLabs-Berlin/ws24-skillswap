```python
import pandas as pd
```

```python
# Create aggregated Dataframe on Factors and dtypes
# Import DataFrames from CSV files
Personal_Info = pd.read_csv('dtype_Personal_Info.csv')
Skill_Exchange = pd.read_csv('dtype_Skill_Exchange.csv')
Track_Record = pd.read_csv('dtype_Track_Record.csv')
Feedback = pd.read_csv('dtype_Feedback.csv')


# Concatenate DataFrames horizontally
combined_df = pd.concat([Personal_Info, Skill_Exchange, Track_Record, Feedback], axis=0, ignore_index=True)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
combined_df.head(100)
```

```python
combined_df.to_csv('dtype_Factors_dtypes.csv', index=False)
```
