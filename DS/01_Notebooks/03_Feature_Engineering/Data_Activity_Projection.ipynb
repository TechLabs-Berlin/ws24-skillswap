```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv(r'C:\Users\Johannes\Documents\ws24-skillswap\DS\01_Notebooks\02_Features_Insights\01_User_Activity_Status\Skillswap_Data_Activity_Status.csv')
pd.set_option('display.max_columns', None)
df1.head()
```

```python
skill_domains = [
    "Agriculture",
    "Animal Care",
    "Art",
    "Beauty",
    "Dance",
    "Music",
    "Writing",
    "Language",
    "Health",
    "Sports",
    "Food",
    "Technology",
    "Photo",
    "Video"
]


filtered_columns = [col for col in df1.columns if col in skill_domains or col == "Sum_Skills_Off" or col == "Sum_Skills_Off_ab_avg"]

filtered_df = df1[filtered_columns]
filtered_df.head()
```

```python
df2 = pd.DataFrame(0, index=df1.index, columns=skill_domains)

df2["Sum_Skills_Off"] = df1["Sum_Skills_Off"].astype(int)
df2["Sum_Skills_Off_ab_avg"] = df1["Sum_Skills_Off_ab_avg"]

for index, row in df1.iterrows():
    for col in df1.columns.drop(["Skill Domains", "Sum_Skills_Off", "Sum_Skills_Off_ab_avg"]):
        if col in row["Skill Domains"]:
            df2.at[index, col] = 1

df2.head()
```

```python
df2["Total_Skills"] = df2[skill_domains].sum(axis=1)

df2.head()
```

```python
sum_skills_off = df2.pop("Sum_Skills_Off")
sum_skills_off_ab_avg = df2.pop("Sum_Skills_Off_ab_avg")


df2["Sum_Skills_Off"] = sum_skills_off
df2["Sum_Skills_Off_ab_avg"] = sum_skills_off_ab_avg
df2.head()
```

```python
df2.to_csv('Data_Activity_Projection.csv', index=False)
```
