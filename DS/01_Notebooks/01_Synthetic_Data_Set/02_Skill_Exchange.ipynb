```python
import pandas as pd
```

```python
# Import DataFrames from CSV files
Skill_Exchange_Offered_uncleaned = pd.read_csv('Skill_Exchange_Offered.csv')
# Drop irrelevant columns
Skill_Exchange_Offered = Skill_Exchange_Offered_uncleaned.drop(columns=["Expert_skill", "Matched_skills_samples"])

Skill_Exchange_Demanded = pd.read_csv('Skill_Exchange_Demanded.csv')


# Concatenate DataFrames horizontally
Skill_Exchange = pd.concat([Skill_Exchange_Offered, Skill_Exchange_Demanded], axis=1)
pd.set_option('display.max_columns', None)
Skill_Exchange.head()
```

```python
Skill_Exchange.to_csv('Skill_Exchange.csv', index=False)
```
