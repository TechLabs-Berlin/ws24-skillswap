```python
Skill_Exchange_Offered = pd.read_csv('Skill_Exchange_Offered.csv')
Skill_Exchange_Offered
```

```python
# Take Expert_Skill dataframe to see where a person can teach (assuming that you can at least teach a skill where you are an expert)
expert_skill = pd.DataFrame(Skill_Exchange_Offered, columns=["Expert_skill"])
expert_skill['Expert_skill'] = expert_skill['Expert_skill'].apply(lambda x: 'Count_' + str(x))
expert_skill
```

```python
concatenated_df = pd.concat([Count_df, expert_skill], axis=1)
concatenated_df
```

```python
# For every user add a random number of lessons taught for his/her expert skill

for index, row in concatenated_df.iterrows():
    expert_skill = row['Expert_skill']
    matching_column = str(expert_skill)
    random_number = max(0, int(np.random.normal(10, 4)))
    concatenated_df.at[index, matching_column] = random_number

concatenated_df
```

```python
concatenated_df_numeric = concatenated_df.apply(pd.to_numeric, errors='coerce')
numeric_columns = concatenated_df_numeric.select_dtypes(include=np.number)
concatenated_df['Sum_Skills_Off'] = numeric_columns.sum(axis=1)
concatenated_df
```

```python
# Create a new column that indicates how many lessons a user has taken, 
# while ensuring that the number of lessons given and taken are balanced


# Step 1: Calculate the sum of the "Sum_Skills_off" column
# Step 2: Generate random numbers for "Sum_Skills_rec" column
# Generate a random number between 80% and 120% of the corresponding value in "Sum_Skills_off" (to ensure a max deviation of 20%)
# Step 3: Adjust the values of the "Sum_Skills_rec" column to match the sum of "Sum_Skills_off"
# Randomly adjust values to match the sum of "Sum_Skills_off" (+ ensure the new value is not negative)


sum_skills_off_sum = concatenated_df['Sum_Skills_Off'].sum()

for index, row in concatenated_df.iterrows():
    min_value = int(0.8 * row['Sum_Skills_Off'])
    max_value = int(1.2 * row['Sum_Skills_Off'])
    if min_value < max_value:
        random_number = np.random.randint(min_value, max_value + 1)  
    else:
        random_number = min_value
    concatenated_df.at[index, 'Sum_Skills_Rec'] = random_number

sum_skills_rec_sum = concatenated_df['Sum_Skills_Rec'].sum()

difference = sum_skills_off_sum - sum_skills_rec_sum
if difference != 0:
    while difference != 0:
        index = np.random.randint(len(concatenated_df))
        current_value = concatenated_df.at[index, 'Sum_Skills_Rec']
        new_value = current_value + np.sign(difference)
        if new_value >= 0:  
            concatenated_df.at[index, 'Sum_Skills_Rec'] = new_value
            difference -= np.sign(difference)

concatenated_df
```

```python
# Sum of skills offered and received must be the same (there is always a teacher and a learner in a skill transaction)
print(concatenated_df['Sum_Skills_Off'].sum())
print(concatenated_df['Sum_Skills_Rec'].sum())

```

```python
skill_domains = [
    "Agriculture",
    "Animal Care",
    "Art",
    "Beauty",
    "Dance",
    "Music",
    "Writing",
    "Language",
    "Health",
    "Sports",
    "Food",
    "Technology",
    "Photo",
    "Video"
]

skills_df = pd.DataFrame(columns=skill_domains)

skills_df.head()
```

```python
concatenated_df2 = pd.concat([concatenated_df, skills_df], axis=1)
concatenated_df2.head()
```

```python
# Skill domains and respective skills
skills_dict = {
    'Agriculture': ['Count_Farming_off', 'Count_Fishery_off', 'Count_Gardening_off'],
    'Animal Care': ['Count_Pet grooming_off', 'Count_Pet training_off'],
    'Art': ['Count_Drawing_off', 'Count_Painting_off', 'Count_Sculpture_off', 'Count_Print making_off', 'Count_Poetry_off'],
    'Beauty': ['Count_Make-up_off', 'Count_Facials_off', 'Count_Pedicure & Manicure_off', 'Count_Hair care_off'],
    'Dance': ['Count_Ballet_off', 'Count_Tap dance_off', 'Count_Jazz dance_off', 'Count_Salsa_off', 'Count_Ballroom dance_off'],
    'Music': ['Count_Piano_off', 'Count_Guitar_off', 'Count_Drumming_off', 'Count_Trumpet_off', 'Count_Composing_off'],
    'Writing': ['Count_Script writing_off', 'Count_Play writing_off', 'Count_Poetry writing_off', 'Count_Caligraphy_off', 'Count_Blogging_off', 'Count_Journalism_off'],
    'Language': ['Count_German_off', 'Count_Spanish_off', 'Count_Arabic_off', 'Count_French_off', 'Count_English_off'],
    'Health': ['Count_Nutrition_off', 'Count_Fitness & Excercise_off', 'Count_Wellness_off'],
    'Sports': ['Count_Tennis_off', 'Count_Football_off', 'Count_Basketball_off', 'Count_Yoga_off', 'Count_Baseball_off', 'Count_Gymnastic_off'],
    'Food': ['Count_Cooking_off', 'Count_Baking_off', 'Count_Grilling_off', 'Count_Meal planning_off', 'Count_Sautéing_off'],
    'Technology': ['Count_Programming_off', 'Count_Graphic design_off', 'Count_Data Analysis_off', 'Count_Photoshop_off', 'Count_Social Media_off'],
    'Photo': ['Count_Camera handling_off', 'Count_Photo editing_off', 'Count_Lighting_off', 'Count_Landscape photography_off', 'Count_Portrait photography_off', 'Count_Image processing_off'],
    'Video': ['Count_Video editing_off', 'Count_Film making_off', 'Count_Animation_off', 'Count_Motion graphics_off', 'Count_Visual effects_off']
}


```

```python
# Based on the expert skill, find out in which skill domain or category the user has this expertise

# Create a new column called "domain"
# Iterate over each row in the DataFrame
# Iterate over each category in the skills_dict
# Check if the expert_skill matches any skill in the category
# Assign the category to the "domain" column
# Exit the loop once a match is found


concatenated_df2['Domain'] = None

for index, row in concatenated_df2.iterrows():
    expert_skill = row['Expert_skill']
    for category, skills in skills_dict.items():
        if any(expert_skill == skill for skill in skills):
            concatenated_df2.at[index, 'Domain'] = category
            break  

pd.set_option('display.max_columns', None)
concatenated_df2.head()
```

```python
# Enter in the column of the skill domain where the user has the expert skill
# how many lessons the user has given in this domain 

# Iterate over each row in the DataFrame
# Check if the domain column exists in the DataFrame
# Copy the value from the "Sum_Skills_off" column to the matched column

for index, row in concatenated_df2.iterrows():
    domain = row['Domain']
    sum_skills_off = row['Sum_Skills_Off']
    
    if domain in concatenated_df2.columns:
        concatenated_df2.at[index, domain] = int(sum_skills_off)

concatenated_df2.head()
```

```python
concatenated_df2.drop(columns=['Expert_skill','Domain'], inplace=True)
concatenated_df2.to_csv('Track_Record.csv', index=False)
```
