```python
import random
import pandas as pd
import numpy as np
```

```python
# Define the skill domains
# Pick a random number between 1 and 3
# Select random skill domains
# Create a list to store the samples
# Generate 750 random skill domain samples


skill_domains = [
    "Agriculture",
    "Animal Care",
    "Art",
    "Beauty",
    "Dance",
    "Music",
    "Writing",
    "Language",
    "Health",
    "Sports",
    "Food",
    "Technology",
    "Photo",
    "Video"
]

def select_random_skills():
    num_skills = random.randint(1, 3)
    selected_skills = random.sample(skill_domains, num_skills)
    return selected_skills

random_skill_samples = []

for _ in range(750):
    random_skills = select_random_skills()
    random_skill_samples.append(random_skills)
    
print(random_skill_samples)
```

```python
Domains_df = pd.DataFrame({"Skill Domains": random_skill_samples})
Domains_df
```

```python
# Skill domains and respective skills
skills_dict = {
    'Agriculture': ['Farming_off', 'Fishery_off', 'Gardening_off'],
    'Animal Care': ['Pet grooming_off', 'Pet training_off'],
    'Art': ['Drawing_off', 'Painting_off', 'Sculpture_off', 'Print making_off', 'Poetry_off'],
    'Beauty': ['Make-up_off', 'Facials_off', 'Pedicure & Manicure_off', 'Hair care_off'],
    'Dance': ['Ballet_off', 'Tap dance_off', 'Jazz dance_off', 'Salsa_off', 'Ballroom dance_off'],
    'Music': ['Piano_off', 'Guitar_off', 'Drumming_off', 'Trumpet_off', 'Composing_off'],
    'Writing': ['Script writing_off', 'Play writing_off', 'Poetry writing_off', 'Caligraphy_off', 'Blogging_off', 'Journalism_off'],
    'Language': ['German_off', 'Spanish_off', 'Arabic_off', 'French_off', 'English_off'],
    'Health': ['Nutrition_off', 'Fitness & Excercise_off', 'Wellness_off'],
    'Sports': ['Tennis_off', 'Football_off', 'Basketball_off', 'Yoga_off', 'Baseball_off', 'Gymnastic_off'],
    'Food': ['Cooking_off', 'Baking_off', 'Grilling_off', 'Meal planning_off', 'Sautéing_off'],
    'Technology': ['Programming_off', 'Graphic design_off', 'Data Analysis_off', 'Photoshop_off', 'Social Media_off'],
    'Photo': ['Camera handling_off', 'Photo editing_off', 'Lighting_off', 'Landscape photography_off', 'Portrait photography_off', 'Image processing_off'],
    'Video': ['Video editing_off', 'Film making_off', 'Animation_off', 'Motion graphics_off', 'Visual effects_off']
}
```

```python
# Initialize a list to store the matched skills for each domain sample
# Iterate through each domain sample in random_skill_samples
# Initialize an empty list to store the matched skills for the current sample
# Iterate through each skill domain in the current sample
# Retrieve the matching skills from skills_dict and extend the matched_skills list
# Append the matched skills for the current sample to the matched_skills_samples list

matched_skills_samples = []

for sample in random_skill_samples:
    matched_skills = []
    for skill_domain in sample:
        matched_skills.extend(skills_dict.get(skill_domain, []))
    matched_skills_samples.append(matched_skills)
    
    
print(matched_skills_samples)
```

```python
Matched_skills_df = pd.DataFrame({"Matched_skills_samples": matched_skills_samples})
Matched_skills_df
```

```python
# Goal: Select a random expert skill
# Initialize a list to store the randomly selected skills
# Iterate over each row in the DataFrame
# Extract the list of matched skills from the current row
# Check if the list of matched skills is not empty
# Select a random skill from the list
# Append the randomly selected skill to the new list
# If the list of matched skills is empty, append None or any placeholder value

expert_skill = []

for index, row in Matched_skills_df.iterrows():
    matched_skills = row['Matched_skills_samples']
    
    if matched_skills:
        random_skill = random.choice(matched_skills)
        expert_skill.append(random_skill)
    else:
        expert_skill.append(None)
```

```python
Expert_df = pd.DataFrame({"Expert_skill": expert_skill})
Expert_df
```

```python
concatenated_df = pd.concat([Skills_Offered, Expert_df, Matched_skills_df, Domains_df], axis=1)
concatenated_df
```

```python
# Make sure that a person is an expert in one skill 
# Extract the value in the "expert_skill" column for the current row
# Iterate over each column in the DataFrame
# Check if the column name matches the expert skill
# Assign "Expert" to the corresponding cell in the matched column

for index, row in concatenated_df.iterrows():
    expert_skill = row['Expert_skill']
    
    for column in concatenated_df.columns:
        if column == expert_skill:
            concatenated_df.at[index, column] = "Expert"
```

```python
# Remove the expert skill from the Matched Skills column to make sure the skill level is not overwritten in the next step
# Iterate over each row in the DataFrame
# Extract the value in the "Expert_Level" column for the current row
# Extract the list of matched skills from the "Matched_skills_samples" column
# Check if the value in the "Expert_Level" column is not null
# Remove the value in "Expert_Level" from the list of matched skills, if it exists
# Update the "Matched_skills_samples" column with the modified list of matched skills


for index, row in concatenated_df.iterrows():
    expert_level = row['Expert_skill']
    
    matched_skills = row['Matched_skills_samples']
    
    if pd.notnull(expert_level):
        if expert_level in matched_skills:
            matched_skills.remove(expert_level)
    
    concatenated_df.at[index, 'Matched_skills_samples'] = matched_skills
```

```python
# Randomly select a value from possible_levels with a normal distribution to 
# fill the columns (where a person has a skill) with skill levels
# Define possible skill levels
# Assumption: Usually people are beginners in the fields where they are not Experts (That's why they are on the platform)
# Iterate over each row in the DataFrame
# Extract the list of matched skills from the current row
# Iterate over each column in the DataFrame
# Check if the column name matches any of the skills in matched_skills
# Define parameters for the normal distribution
# Generate random index from a normal distribution
# Ensure the index falls within the range of possible indices
# Select the skill level corresponding to the generated index
# Assign the randomly selected value to the corresponding cell in the matched headline


possible_levels = ["Novice", "Beginner", "Intermediate", "Advanced", "Expert"]

average_range = [possible_levels.index("Beginner"), possible_levels.index("Intermediate")]

for index, row in concatenated_df.iterrows():
    matched_skills = row['Matched_skills_samples']
    
    for column in concatenated_df.columns:
        if column in matched_skills:
            mean = np.mean(average_range)  
            std_dev = len(possible_levels) / 3  
                        
            random_index = int(np.random.normal(mean, std_dev))            
            random_index = max(0, min(random_index, len(possible_levels) - 1))
            random_value = possible_levels[random_index]
            concatenated_df.at[index, column] = random_value
```
