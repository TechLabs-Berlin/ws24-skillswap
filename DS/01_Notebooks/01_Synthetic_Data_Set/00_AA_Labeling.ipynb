```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Johannes\Documents\Uni\Master\Tech_Labs\ws24-skillswap\DS\data\01_raw\Skillswap_Data_Raw.csv')

pd.set_option('display.max_columns', None)

df.head()
```

```python
# Label Data with Activity Status

# 1. Make a Column Called "Activity_Status"
df['Activity_Status'] = ""

# 2. Iterate through the rows and apply conditions
average_sum_skills = df['Sum_Skills_Off'].mean()
df['Sum_Skills_Off_ab_avg'] = df['Sum_Skills_Off'].apply(lambda x: 1 if x > average_sum_skills else 0)

for index, row in df.iterrows():
    if row['Last_Seen'] in ["Recently", "Today", "Yesterday", "Within a Week"] and \
            row['Response_Time'] in ["Within 24 hours", "Within a week"] and \
            row['Sum_Skills_Off_ab_avg'] == 1:
        df.at[index, 'Activity_Status'] = "Swapper"
    elif row['Last_Seen'] in ["Recently", "Today", "Yesterday", "Within a Week"] and \
            row['Response_Time'] in ["Within 24 hours", "Within a week"] and \
            row['Sum_Skills_Off_ab_avg'] == 0:
        df.at[index, 'Activity_Status'] = "Active User"
    elif row['Last_Seen'] == "Within a Month" or row['Response_Time'] == "Within a month":
        df.at[index, 'Activity_Status'] = "Occasional User"
    elif row['Last_Seen'] == "Over a Month Ago" or row['Response_Time'] == "More than a month":
        df.at[index, 'Activity_Status'] = "Inactive"
    else:
        df.at[index, 'Activity_Status'] = pd.NA 

df.head()
```

```python
# Check NA Values
empty_values = df['Activity_Status'].isna().sum()

print("Number of empty values in the 'Activity_Status' column:", empty_values)
```

```python
# Count unique values
activity_status_counts = df['Activity_Status'].value_counts()

print("Count of unique values in the 'Activity_Status' column:")
print(activity_status_counts)
```

```python
df.to_csv('Skillswap_Data_Labeled.csv', index=False)
```
